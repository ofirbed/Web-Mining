```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

wiki_url = 'https://en.wikipedia.org/wiki/Julia_Roberts_filmography'
page = urlopen(wiki_url)
soup = BeautifulSoup(page)
table=soup.find('table', class_='wikitable plainrowheaders sortable')
year=[]
title=[]
role=[]
directors=[]

for row in table.findAll("tr"):
    cells = row.findAll('td')
    header=row.findAll('th') 
    if len(cells)==5:
        year.append(cells[0].find(text=True).strip('\n'))
        title.append(header[0].find(text=True).strip('\n'))
        role.append(cells[1].find(text=True).strip('\n'))
        directors.append(cells[2].find(text=True).strip('\n'))
       

df=pd.DataFrame(year,columns=['Year'])
df['Title']=title
df['Role']=role
df['Directors']=directors

df
        
```
